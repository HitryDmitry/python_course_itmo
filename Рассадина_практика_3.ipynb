```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/titanic/train.csv')
```

```python
# Data types
df.info()
```

```python
# Number of nans
df.isna().sum()
```

```python
# Describe df
df.describe()
```

```python
# Процент выживаемости у каждого класса пассажиров
df.groupby('Pclass')['Survived'].mean()
```

```python
# Выведите самое популярное мужское и самое популярное женское имя на корабле
# Выведем имена мужчин
df[df['Sex'] == 'male']['Name'].value_counts()
```

```python
# Имена расположены в порядке убывания встречаемости сверху вниз, поэтому нам остается выбрать самое верхнее имя (с нулевым индексом)
print(df[df['Sex'] == 'male']['Name'].value_counts().index[0])
print(df[df['Sex'] == 'female']['Name'].value_counts().index[0])
```

```python
# Выведите часть таблицы с пассажирами, возраст которых больше 44 лет
df[df['Age'] > 44]
```

```python
# Выведите часть таблицы с пассажирами, возраст которых меньше 44 лет и которые мужского пола
df[(df['Age'] < 44) & (df['Sex'] == 'male')]
```

```python
# Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека)
# Посчитаем, сколько человек было в каждой кабине
counts = df['Cabin'].value_counts()
counts
```

```python
# Выведем количества n-местных кабин
print(counts.value_counts()) # С 1-местными кабинами
print("------------------------------")
print(counts.value_counts()[1:]) # n-местные
```
