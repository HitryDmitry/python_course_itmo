```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('../datasets/titanic/train.csv')
```

### Визуализируйте распределение значений признаков Survived, Pclass, Age, Sex, Parch

```python
features = ['Survived', 'Pclass', 'Age', 'Sex', 'Parch']
ax, fig = plt.subplots()

for f in features:
    ax = sns.histplot(df[f], kde=True)
    plt.title(f"Распределение значений призкака {f}")
    plt.show()
```

### Постройте график типа boxplot для столбца Age

```python
sns.boxplot(df['Age'])
plt.title("График типа Boxplot для Age")
plt.show()
```

Проинтепретируйте полученный график

Ответ: Минимальный возраст - 1-2 года, максимальный - 62-63 (без учета выбросов). Медиана - примерно 28 лет (т.е. половина пассажиров младше 28, а другая - старше). Первый квартиль - 20, третий - 38. Также присутствуют выбросы (пассажиры с возрастом больше 65).

### Постройте график типа pie chart для переменных Survived, Pclass, подпишите доли в процентах

```python
df['Survived'].value_counts().plot(kind='pie', subplots=True, autopct='%1.1f%%', startangle=270, title="Соотношение количества выживших и невыживших")
```

```python
df['Pclass'].value_counts().plot(kind='pie', subplots=True, autopct='%1.1f%%', startangle=270, title="Соотношение количества мужчин и женщин")
```

### Постройте график типа pairplot для всех числовых переменных датасета

Посмотрим на типы признаков

```python
df.info()
```

```python
df.columns[(df.dtypes == np.int64) | (df.dtypes == np.float64)]
```

```python
# Используем функцию pairplot seaborn. vars - переменные, которые мы будем отображать
pfig = sns.pairplot(df, vars=df.columns[(df.dtypes == np.int64) | (df.dtypes == np.float64)])
pfig.fig.suptitle("Pairplot для числовых признаков", y=1.05)
plt.show()
```

### Постройте интерактивный sunburst plot (визуализация иерархических данных) с помощью plotly. На первом уровне иерархии - количество пассажиров в каждом из классов, а на втором количество женщин/мужчин в этом классе

```python
fig = px.sunburst(df, path=['Pclass', 'Sex'])
fig.update_layout(title_text="Интерактивный sunburst plot для признаков Pclass, Sex", title_x = 0.5)
fig.show()
```
